```python
import pandas as pd
import numpy as np
```

```python
total_gdp_df = pd.read_csv('total_gdp.csv')
```

```python
total_gdp_df.head()
```

```python
total_gdp_df.drop(['INDICATOR', 'SUBJECT', 'MEASURE', 'FREQUENCY', 'Flag Codes'], axis=1, inplace=True)
```

```python
import duckdb
```

```python
total_gdp_df.head()
```

```python
total_gdp_df = duckdb.sql("SELECT LOCATION AS Country, TIME AS Year, Value AS 'Total_GDP_Million_USD' FROM total_gdp_df").df()
```

```python
total_gdp_df.head()
```

```python
print(total_gdp_df['Country'].unique())
```

```python
total_gdp_df = total_gdp_df[total_gdp_df['Country'] != 'OECD']
```

```python
print(total_gdp_df['Country'].unique())
```

```python
total_gdp_df
```

```python
oecd_countries = ["AUS", "AUT", "BEL", "CAN", "CHL", "CZE", "DNK", "EST", "FIN", "FRA", "DEU",
    "GRC", "HUN", "ISL", "IRL", "ISR", "ITA", "JPN", "KOR", "LVA", "LTU", "LUX",
    "MEX", "NLD", "NZL", "NOR", "POL", "PRT", "SVK", "SVN", "ESP", "SWE", "CHE",
    "TUR", "GBR", "USA"]
```

```python
years = np.arange(1960, 2023)
data = []
for country in oecd_countries:
    for year in years:
        data.append([country, year])
skeleton_df = pd.DataFrame(data, columns=['Country', 'Year'])
skeleton_df
```

```python
df = duckdb.sql("SELECT skeleton_df.Country, skeleton_df.Year, total_gdp_df.Total_GDP_Million_USD\
                           FROM skeleton_df LEFT JOIN total_gdp_df\
                           ON skeleton_df.Country = total_gdp_df.Country\
                           AND skeleton_df.Year = total_gdp_df.Year").df()
```

```python
df
```

```python
df = duckdb.sql("SELECT * FROM df ORDER BY Country, Year").df()
```

```python
df
```
